## 4. РљРѕРјР±РёРЅРёСЂРѕРІР°РЅРЅС‹Рµ РјРµС‚РѕРґС‹  

- **РџРѕСЃР»РѕРіРѕРІР°СЏ С‚РѕРєРµРЅРёР·Р°С†РёСЏ** (Р°РєС‚СѓР°Р»СЊРЅРѕ РґР»СЏ СЏР·С‹РєРѕРІ СЃ СЏРІРЅРѕР№ СЃР»РѕРіРѕРІРѕР№ СЃС‚СЂСѓРєС‚СѓСЂРѕР№, РЅР°РїСЂРёРјРµСЂ, СЏРїРѕРЅСЃРєРёР№)

## 5. РџСЂРёРјРµСЂ РїСЂРµРґРѕР±СЂР°Р±РѕС‚РєРё С‚РµРєСЃС‚Р°

Р’РѕР·СЊРјС‘Рј РЅРѕРІРѕСЃС‚РЅРѕР№ С‚РµРєСЃС‚ Рё РїСЂРёРјРµРЅРёРј РїРѕР»РЅС‹Р№ РїР°Р№РїР»Р°Р№РЅ РїСЂРµРґРѕР±СЂР°Р±РѕС‚РєРё: РѕС‡РёСЃС‚РєР° в†’ С‚РѕРєРµРЅРёР·Р°С†РёСЏ в†’ Р»РµРјРјР°С‚РёР·Р°С†РёСЏ в†’ СѓРґР°Р»РµРЅРёРµ СЃС‚РѕРї-СЃР»РѕРІ


```python
import spacy
from nltk.corpus import stopwords
import nltk
import re

nltk.download('stopwords')

# Р—Р°РіСЂСѓР·РєР° РјРѕРґРµР»Рё spaCy РґР»СЏ Р°РЅРіР»РёР№СЃРєРѕРіРѕ
nlp = spacy.load("en_core_web_sm")
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # 1. РџСЂРёРІРµРґРµРЅРёРµ Рє РЅРёР¶РЅРµРјСѓ СЂРµРіРёСЃС‚СЂСѓ
    text = text.lower()
    # 2. РЈРґР°Р»РµРЅРёРµ РїСѓРЅРєС‚СѓР°С†РёРё Рё С†РёС„СЂ
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # 3. РўРѕРєРµРЅРёР·Р°С†РёСЏ Рё Р»РµРјРјР°С‚РёР·Р°С†РёСЏ С‡РµСЂРµР· spaCy
    doc = nlp(text)
    # 4. РЈРґР°Р»РµРЅРёРµ СЃС‚РѕРї-СЃР»РѕРІ Рё РєРѕСЂРѕС‚РєРёС… С‚РѕРєРµРЅРѕРІ
    tokens = [token.lemma_ for token in doc if token.text not in stop_words and len(token.text) > 2]
    return tokens

# РџСЂРёРјРµСЂ
sample_text = "The quick brown fox jumps over the lazy dog. This is a simple sentence for preprocessing!"
processed = preprocess_text(sample_text)
print("РћСЂРёРіРёРЅР°Р»:", sample_text)
print("РџРѕСЃР»Рµ РїСЂРµРґРѕР±СЂР°Р±РѕС‚РєРё:", processed)
```

```python
# РўРµРїРµСЂСЊ РїРѕРїСЂРѕР±СѓР№С‚Рµ СЃР°РјРё:
# 1. РќР°РїРёС€РёС‚Рµ СЃРІРѕСЋ С„СѓРЅРєС†РёСЋ РїСЂРµРґРѕР±СЂР°Р±РѕС‚РєРё, РёСЃРїРѕР»СЊР·СѓСЏ Р»СЋР±РѕР№ РёР· РёР·СѓС‡РµРЅРЅС‹С… С‚РѕРєРµРЅРёР·Р°С‚РѕСЂРѕРІ
# 2. РџСЂРёРјРµРЅРёС‚Рµ РµС‘ Рє С‚РµРєСЃС‚Сѓ: "Natural Language Processing is a fascinating field of Artificial Intelligence!"
# 3. Р’С‹РІРµРґРёС‚Рµ СЂРµР·СѓР»СЊС‚Р°С‚

### РІР°С€ РєРѕРґ Р·РґРµСЃСЊ ###
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import remove_stopwords
import re

def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # СѓРґР°Р»РµРЅРёРµ СЃС‚РѕРї-СЃР»РѕРІ
    filtered_text = remove_stopwords(text)
    return simple_preprocess(text)

# РџСЂРёРјРµСЂ
sample_text = "Natural Language Processing is a fascinating field of Artificial Intelligence!"
processed = preprocess_text(sample_text)
print("РћСЂРёРіРёРЅР°Р»:", sample_text)
print("РџРѕСЃР»Рµ РїСЂРµРґРѕР±СЂР°Р±РѕС‚РєРё:", processed)
```
